```python
# 4_3_Analisis_Chart_Manual.py
# Grafik perbandingan performa Rule-Based vs Random Forest
# pada 50 vs 100 topologi (Precision, Recall, F1, Accuracy)

import matplotlib.pyplot as plt

# ==== DATA MANUAL DARI HASIL UJI ====
x = [50, 100]  # jumlah topologi

rule_based = {
    "Precision": [1.0000, 1.0000],
    "Recall":    [1.0000, 1.0000],
    "F1":        [1.0000, 1.0000],
    "Accuracy":  [1.0000, 1.0000]
}

random_forest = {
    "Precision": [0.6800, 0.7600],
    "Recall":    [0.5707, 0.6554],
    "F1":        [0.6037, 0.6905],
    "Accuracy":  [0.9225, 0.9425]
}

# ==== PLOT 4 SUBPLOT ====
fig, axes = plt.subplots(1, 4, figsize=(16, 4), constrained_layout=True)
metrics = ["Precision", "Recall", "F1", "Accuracy"]

for ax, m in zip(axes, metrics):
    ax.plot(x, rule_based[m], marker="o", color="tab:blue", label="Rule-Based")
    ax.plot(x, random_forest[m], marker="s", color="tab:orange", label="Random Forest")
    ax.set_title(m)
    ax.set_xlabel("Jumlah Topologi")
    ax.set_ylabel(m)
    ax.set_xticks(x)
    ax.set_ylim(0, 1.05)
    ax.grid(True, alpha=0.3)

    # tampilkan nilai di atas titik
    for xi, yi in zip(x, rule_based[m]):
        ax.text(xi, yi + 0.02, f"{yi:.2f}", ha='center', color="tab:blue", fontsize=9)
    for xi, yi in zip(x, random_forest[m]):
        ax.text(xi, yi + 0.02, f"{yi:.2f}", ha='center', color="tab:orange", fontsize=9)

# legend hanya sekali di panel pertama
axes[0].legend()
fig.suptitle("Perbandingan Kinerja Rule-Based dan Random Forest\n(50 vs 100 Topologi)", fontsize=13)
plt.savefig("bab4_analisis_perbandingan_manual.png", dpi=200)
plt.show()
```
